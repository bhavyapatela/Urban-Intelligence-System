```python
# New Column for Weekend or Weekday
df["day_type"]= df["is_weekend"].map({0: "Weekday", 1: "Weekend"})
df.head()
```

```python
#PEAK TRAFFIC HOURS
hourly_traffic= df.groupby("hour")["congestion_index"].mean()

plt.figure()
hourly_traffic.plot()
plt.title("Average Congestion by Hour")
plt.xlabel("Hour")
plt.ylabel("Congestion Index")
plt.show()

# Insight: Congestion levels peak during morning (8–10 AM) and evening (6–8 PM) hours, indicating typical office commute patterns and high road usage during these periods.
```

```python
# Day-wise Traffic Pattern

sns.barplot(x="day_of_week", y="congestion_index", data=df)
plt.title("Traffic by Day of Week")
plt.xlabel("Day Of Week")
plt.ylabel("Congestion Index")
plt.show()

#Insight: Traffic congestion remains consistently high during weekdays (days 0–4) and drops significantly on weekends (days 5–6), indicating that work-related travel is a major contributor to urban traffic.
```

```python
# Weekend vs Weekday
sns.boxplot(x="day_type", y="congestion_index", data=df)
plt.title("Weekend vs Weekday Traffic")
plt.xlabel("Day Type")
plt.ylabel("Congestion Index")
plt.show()

# Insight: Weekdays show significantly higher congestion compared to weekends, suggesting that work-related travel is a major contributor to urban traffic.
```

```python
# Pollution Trends (AQI vs Hour)
aqi_hour = df.groupby("hour")["aqi"].mean()

plt.figure()
aqi_hour.plot(color='red')
plt.title("AQI by Hour")
plt.xlabel("Hour of the day")
plt.ylabel("AQI")
plt.show()

#Insight: AQI levels show a steady increase throughout the day, with the highest values observed during evening hours. This suggests that air pollution accumulates over time, likely due to increased human activities and traffic during the day.
```

```python
# Traffic vs Pollution (MOST IMPORTANT)

sns.scatterplot(x="congestion_index", y="aqi", data=df)
plt.title("Traffic vs AQI")
plt.xlabel("Congestion Index")
plt.ylabel("AQI")
plt.show()

#Insight: While some high congestion instances correspond to elevated AQI levels, the overall distribution is scattered, indicating that traffic alone is not the sole determinant of air pollution. Other environmental factors likely play a significant role.
```

```python
# Correlation Heatmap 
corr = df[['congestion_index','aqi','pm25','pm10','no2','co','temperature','humidity']].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Weather Impact on Traffic
sns.scatterplot(x="precipitation", y="congestion_index", data=df)
plt.title("Rain vs Traffic")
plt.xlabel("Precipitation")
plt.ylabel("Congestion Index")
plt.show()

#Insight: Most observations occurred under no-rain conditions, where congestion levels vary significantly due to other factors such as time of day or traffic demand. Due to very limited data during rainfall, no strong relationship between precipitation and congestion can be conclusively established.
```

```python
#Visibility Impact

sns.scatterplot(x="visibility", y="congestion_index", data=df)
plt.title("Visibility vs Traffic")
plt.show()

# Insight: The majority of observations occur under high visibility conditions, where congestion levels still vary significantly. Due to the lack of sufficient low-visibility data points, no clear relationship between visibility and traffic congestion can be established.
```

```python
# Heatmap (Hour vs Day) — VERY IMPORTANT
pivot = df.pivot_table(values="congestion_index", index="day_of_week", columns="hour")

sns.heatmap(pivot, cmap="YlOrRd")
plt.title("Traffic Heatmap (Day vs Hour)")
plt.xlabel("Hour")
plt.ylabel("Day of Week")
plt.show()
```

```python
#Combine Insight 

sns.scatterplot(x="temperature", y="aqi", hue="congestion_index", data=df)
plt.title("Weather + Traffic + AQI")
plt.show()
```
